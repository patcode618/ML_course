```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

#Create your df here:
df = pd.read_csv("profiles.csv")
print(df)
```

```python
df.dropna(subset=["height","income","job",
                   "essay0","essay1","essay2","essay3","essay4","essay5","essay6","essay7","essay8","essay9",
                   "body_type","diet", "drinks", "drugs","education","last_online","offspring","orientation","sex", "status",
                   "location","pets","religion","sign", "smokes", "speaks", "ethnicity"],
           inplace=True)
df.reset_index(drop=True)
```

```python
drink_mapping = {"not at all": 0, "rarely": 1, "socially": 2, "often": 3, "very often": 4, "desperately": 5}
df["drinks_code"] = df.drinks.map(drink_mapping)

smokes_mapping = {"no": 0, "yes":1, "sometimes":2, "when drinking": 3, "trying to quit": 4}
df["smokes_code"] = df.smokes.map(smokes_mapping)

drugs_mapping = {"never": 0, "sometimes": 1, "often": 2}
df["drugs_code"] = df.drugs.map(drugs_mapping)

df.reset_index(drop=True)
# ADD others if wanted
```

```python
# Clean the data
import re
def preprocessor(text):
    text = re.sub('<[^>]*>', '', text)
    emoticons = re.findall('(?::|;|=) (?:-)?(?:\) |\(|D|P)',
                           text)
    text = (re.sub('[\W]+', ' ', text.lower()) +
           ' '.join(emoticons).replace('-', ''))
    return text

# Apply it; to clean it
df['essay0'] = df['essay0'].apply(preprocessor)
df['essay1'] = df['essay1'].apply(preprocessor)
df['essay2'] = df['essay2'].apply(preprocessor)
df['essay3'] = df['essay3'].apply(preprocessor)
df['essay4'] = df['essay4'].apply(preprocessor)
df['essay5'] = df['essay5'].apply(preprocessor)
df['essay6'] = df['essay6'].apply(preprocessor)
df['essay7'] = df['essay7'].apply(preprocessor)
df['essay8'] = df['essay8'].apply(preprocessor)
df['essay9'] = df['essay9'].apply(preprocessor)
```

```python
# work with clean data
essay_cols = ["essay0","essay1","essay2","essay3","essay4","essay5","essay6","essay7","essay8","essay9"]

all_essays = df[essay_cols].apply(lambda x: ' '.join(x), axis=1)

df["essay_len"] = all_essays.apply(lambda x: len(x))

#print(all_essays) # cool
df_all_essays = pd.DataFrame(all_essays, columns=['essay']).reset_index(drop=True)
df_all_essays
```

```python
# Bag-of-words for LDA -- count = vectorizer
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english',
                       max_df=.1,
                       max_features=5000)
X = count.fit_transform(all_essays.values)
print(X)

from sklearn.decomposition import LatentDirichletAllocation

lda = LatentDirichletAllocation(n_components=10,
                               random_state=123,
                               learning_method='batch',
                               learning_decay=.5)
X_topics = lda.fit_transform(X)
print(X_topics)
```

```python
lda.components_.shape
```

```python
n_top_words = 5 # 5 words explanation
feature_names = count.get_feature_names()

for topic_idx, topic in enumerate(lda.components_):
    print("Topic %d:" % (topic_idx + 1))
    print(" ".join([feature_names[i]
                    for i in topic.argsort()\
                    [:-n_top_words - 1:-1]]))
# each of those topics gets a bit fine tuned (by person) to make more sense - as seen in the confirmation cell
```

```python
# confirm the categories 
east = X_topics[:, 0].argsort()[::-1]

for iter_idx, profiles_idx in enumerate(east[:3]):
    print('\nQuestion Hate Wish #%d:' % (iter_idx +1))
    print(df_all_essays['essay'][profiles_idx][:1000], "...")
```

```python
# Diagnose model performance with log-likelihood
print("Log Likelihood: ", lda.score(X)) # higher the better
print("Perplexity: ", lda.perplexity(X))
print(lda.get_params())
```

```python
# Getting the params
from sklearn.model_selection import GridSearchCV

search_params = {'n_components': [10, 15, 20, 30], 'learning_decay': [.5, .7, .9]}

lda_model = LatentDirichletAllocation()  # init the model

model = GridSearchCV(lda_model, param_grid=search_params) # init GridSearch Class

model.fit(X) # do GridSearch
```

```python
# Best Model
best_lda_model = model.best_estimator_

# Model Parameters
print("Best Model's Params: ", model.best_params_)

# Log Likelihood Score
print("Best Log Likelihood Score: ", model.best_score_)
```

```python
# Get Log Likelyhoods from Grid Search Output
n_components = [10, 15, 20, 30]
log_likelyhoods_5 = [round(gscore.mean_validation_score) for gscore in model.grid_scores_ if gscore.parameters['learning_decay']==0.5]
log_likelyhoods_7 = [round(gscore.mean_validation_score) for gscore in model.grid_scores_ if gscore.parameters['learning_decay']==0.7]
log_likelyhoods_9 = [round(gscore.mean_validation_score) for gscore in model.grid_scores_ if gscore.parameters['learning_decay']==0.9]

# Show graph
plt.figure(figsize=(12, 8))
plt.plot(n_components, log_likelyhoods_5, label='0.5')
plt.plot(n_components, log_likelyhoods_7, label='0.7')
plt.plot(n_components, log_likelyhoods_9, label='0.9')
plt.title("Choosing Optimal LDA Model")
plt.xlabel("Num Components")
plt.ylabel("Log Likelyhood Scores")
plt.legend(title='Learning decay', loc='best')
plt.show()
```

```python
# Topic Matrix
lda_output = best_lda_model.transform(X)

# column names
topicnames = ["Topic" + str(i) for i in range(best_lda_model.n_components)]

# index names
docnames = ["Essay" + str(i) for i in range(len(df_all_essays))]

# mae it df
df_document_topic = pd.DataFrame(np.round(lda_output, 2), columns=topicnames, index=docnames)

# Get dominant topic
dominant_topic = np.argmax(df_document_topic.values, axis=1)
df_document_topic['dominant_topic'] = dominant_topic


def color_red(val):
    color = 'red' if val > .1 else 'black'
    return 'color: {col}'.format(col=color)

def make_bold(val):
    weight = 700 if val > .1 else 400
    return 'font-weight: {weight}'.format(weight=weight)

# Apply Style
df_document_topics = df_document_topic.head(15).style.applymap(color_red).applymap(make_bold)
df_document_topics
```

```python
# top n_words for topic
def show_topics(vectorizer=count, lda_model=lda_model, n_words=20):
    keywords = np.array(count.get_feature_names())
    topic_keywords = []
    for topic_weights in lda.components_:
        top_keyword_locs = (-topic_weights).argsort()[:n_words]
        topic_keywords.append(keywords.take(top_keyword_locs))
    return topic_keywords

topic_keywords = show_topics(vectorizer=count, lda_model=best_lda_model, n_words=15)        

df_topic_keywords = pd.DataFrame(topic_keywords)
df_topic_keywords.columns = ['Word '+str(i) for i in range(df_topic_keywords.shape[1])]
df_topic_keywords.index = ['Topic '+str(i) for i in range(df_topic_keywords.shape[0])]
df_topic_keywords # shows n_words for topic 
```
